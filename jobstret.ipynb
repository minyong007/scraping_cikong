```python
import requests
from bs4 import BeautifulSoup
import csv

jobs_data = []

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:147.0) Gecko/20100101 Firefox/147.0"
}

s = requests.session()
s.headers.update(headers)

i = 1  # mulai dari halaman 1
while True:
    url = f'https://id.jobstreet.com/id/data-scientist-jobs?page={i}&sortmode=ListedDate'
    try:
        page = s.get(url)
        soup = BeautifulSoup(page.text, "lxml")

        artikel_list = soup.find_all('article', attrs={'data-automation':'normalJob'})
        if not artikel_list:  # kalau sudah tidak ada artikel, hentikan loop
            break

        for artikel in artikel_list:
            posisi = artikel.find(attrs={'data-automation':'jobTitle'})
            perusahaan = artikel.find(attrs={'data-automation':'jobCompany'})
            lokasi = artikel.find(attrs={'data-automation':'jobLocation'})
            gaji = artikel.find(attrs={'data-automation':'jobSalary'})
            waktu = artikel.find(attrs={'data-automation':'jobListingDate'})

            posisi_text = posisi.get_text() if posisi else "Posisi tidak ditemukan"
            perusahaan_text = perusahaan.get_text() if perusahaan else "Perusahaan tidak ditemukan"
            lokasi_text = lokasi.get_text() if lokasi else "Lokasi tidak ditemukan"
            gaji_text = gaji.get_text() if gaji else "Gaji tidak disebutkan"
            waktu_text = waktu.get_text() if waktu else "Waktu tidak ditemukan"

            print(posisi_text, perusahaan_text, lokasi_text, gaji_text, waktu_text, sep="\n")
            print("==========================================")

            jobs_data.append({
                "Posisi": posisi_text,
                "Perusahaan": perusahaan_text,
                "Lokasi": lokasi_text,
                "Gaji": gaji_text,
                "Waktu": waktu_text
            })

        print(f"Halaman {i} selesai di-scrape.")

        i += 1  # lanjut ke halaman berikutnya

    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        break
```

```python
df_jobs = pd.DataFrame(jobs_data)
df_jobs.to_csv('[Jobstreet]Data_Job_Listings.csv', index=False)
```
